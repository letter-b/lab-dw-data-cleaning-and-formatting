## Exercise 1: Cleaning Column Names

To ensure consistency and ease of use, standardize the column names of the dataframe. Start by taking a first look at the dataframe and identifying any column names that need to be modified. Use appropriate naming conventions and make sure that column names are descriptive and informative.

*Hint*:

```python
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
df = pd.read_csv(url)
df.head()

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df.columns

df.rename(columns={'st':'state'}, inplace=True)
df.columns

```

## Exercise 2: Cleaning invalid Values

The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis. Therefore, we need to clean these columns to ensure that they only contain valid data.

Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).

*Hint*:
- *Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*
- *State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*
- *In education, "Bachelors" could be replaced by "Bachelor"*
- *In Customer Lifetime Value, delete the `%` character*
- *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

```python
#Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".
print(df['gender'].unique())
df['gender'] = df['gender'].replace({'Femal':'F', 'female':'F', 'Male':'M'})
```

```python
#State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"
print(df['state'].unique())
df['state'] = df['state'].replace({'AZ':'Arizona', 'Cali':'California', 'WA':'Washington'})
```

```python
#In education, "Bachelors" could be replaced by "Bachelor"
df['education'].unique()
df['education'] = df['education'].replace({'Bachelors':'Bachelor'})
```

```python
#In Customer Lifetime Value, delete the % character
df['customer_lifetime_value'] = df['customer_lifetime_value'].str.replace('%', '')
```

```python
#In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"
print(df['vehicle_class'].unique())
df['vehicle_class'] = df['vehicle_class'].replace({'Sports Car':'Luxury', 'Luxury SUV':'Luxury', 'Luxury Car':'Luxury'})
```

```python
# Identifying null values
df.isnull().sum()
```

```python
# Drop the rows or columns with null values
df = df.dropna() # Removed this line as it caused the DataFrame to become empty
```

```python
# Find the type per column
df.dtypes
```

```python
#Fill the null values with a specific value (such as the column mean or median for numerical variables, and mode for categorical variables)
df['customer_lifetime_value'] = df['customer_lifetime_value'].fillna(df['customer_lifetime_value'].mean())
df['income'] = df['income'].fillna(df['income'].median())
df['monthly_premium_auto'].fillna(df['monthly_premium_auto'].median())
df['number_of_open_complaints'] = df['number_of_open_complaints'].fillna(df['number_of_open_complaints'].median())
df['total_claim_amount'] = df['total_claim_amount'].fillna(df['total_claim_amount'].median())

df['customer'] = df['customer'].fillna(df['customer'].mode()[0])
df['state'] = df['state'].fillna(df['state'].mode()[0])
df['gender'] = df['gender'].fillna(df['gender'].mode()[0])
df['education'] = df['education'].fillna(df['education'].mode()[0])
df['policy_type'] = df['policy_type'].fillna(df['policy_type'].mode()[0])
df['vehicle_class'] = df['vehicle_class'].fillna(df['vehicle_class'].mode()[0])
```

```python
#MUCH SIMPLER WAY:
# Loop over each column and fill missing values based on its data type
for column in df.columns:
    if df[column].dtype in [np.float64]:  # Numerical columns
        mean_value = df[column].mean()
        df[column] = df[column].fillna(mean_value)
    elif df[column].dtype == object:  # Categorical columns
        mode_value = df[column].mode()[0]
        df[column] = df[column].fillna(mode_value)

# Verifying the handling of null values
null_counts_after = df.isnull().sum()
print(f"Columns with null values after handling:\n{null_counts_after[null_counts_after > 0]}")
```

```python
#Fill the null values with the previous or next value in the column
df = df.ffill()
print(df.head())
```

```python
df.isnull().sum()

# Approach: Re-checked the dataset using isnull().sum() to ensure all null values were appropriately handled.
# Outcome: No null values remained, indicating successful execution of the strategy.

#Reasoning:
#Numerical Columns: Chosen to fill missing values with the mean because it maintains the average trend and is less sensitive to outliers than the median.
#Categorical Columns: Chosen to fill missing values with the mode, as it is the most common occurrence in the data, thereby preserving the distribution and variability of the data.

#Columns such as customer_lifetime_value, income, monthly_premium_auto, number_of_open_complaints, and total_claim_amount had missing values (specific null counts can be added).
#Categorical columns such as customer, state, gender, education, etc., were also checked, although they typically have fewer missing values.
```

## **Bonus: Challenge 2: creating functions on a separate `py` file**

Put all the data cleaning and formatting steps into functions, and create a main function that performs all the cleaning and formatting.

Write these functions in separate .py file(s). By putting these steps into functions, we can make the code more modular and easier to maintain.

*Hint: autoreload module is a utility module in Python that allows you to automatically reload modules in the current session when changes are made to the source code. This can be useful in situations where you are actively developing code and want to see the effects of changes you make without having to constantly restart the Python interpreter or Jupyter Notebook kernel.*

```python
import pandas as pd

def load_data(url):
    return pd.read_csv(url)

def handle_missing_values(df):
    # Example: fill NA with the mean of the column
    return df.fillna(df.mean())

def remove_duplicates(df):
    return df.drop_duplicates()

def rename_columns(df):
    # Example: make all column names lowercase
    df.columns = df.columns.str.lower()
    return df

def clean_data(url):
    df = load_data(url)
    df = handle_missing_values(df)
    df = remove_duplicates(df)
    df = rename_columns(df)
    return df
```
